## Outlier detection and Removal 

```python
#Plot the distribution plots for the 'Fare' feature
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(10,5))
plt.subplot(1,1,1)
sns.distplot(df['Fare'])
plt.show()
```

```python
#boxplot for 'Fare' Feature
plt.boxplot(df['Fare'])
plt.show()
```

```python
#Finding IQR
percentile25 = df['Fare'].quantile(0.25)
percentile75 = df['Fare'].quantile(0.75)
iqr=percentile75-percentile25
#finding upper and lower limits
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
#finding outliers
df[df['Fare'] > upper_limit]
df[df['Fare'] < lower_limit]
#trimming outliers
new_df = df[df['Fare'] < upper_limit]
```

```python
# Comparing plots after trimmimg
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['Fare'])
plt.subplot(2,2,2)
sns.boxplot(df['Fare'])
plt.subplot(2,2,3)
sns.distplot(new_df['Fare'])
plt.subplot(2,2,4)
sns.boxplot(new_df['Fare'])
plt.show()
```

```python
#Capping
new_df_cap = df.copy()

new_df_cap['Fare'] = np.where(
    new_df_cap['Fare'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['Fare'] < lower_limit,
        lower_limit,
        new_df_cap['Fare']
    )
)
```

```python
#Comparing plots after capping
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['Fare'])
plt.subplot(2,2,2)
sns.boxplot(df['Fare'])
plt.subplot(2,2,3)
sns.distplot(new_df_cap['Fare'])
plt.subplot(2,2,4)
sns.boxplot(new_df_cap['Fare'])
plt.show()
```
